```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

stocks = ['1454', '2539', '4438', '3617', '6139']

for i in stocks:
    df = pd.read_csv('yahoo' + i + 'TW.csv')
    df.set_index('Date', inplace=True)

    # 進場策略
    df['MA50'] = df['Close'].rolling(window=50).mean()
    df['MA200'] = df['Close'].rolling(window=200).mean()
    df['Trend'] = np.where(df['MA50'] > df['MA200'], 1, -1)
    df['RSI'] = df['Close'].rolling(window=14).apply(lambda x: x.mean() / x.std())
    df['Buy_Signal'] = np.where((df['Trend'] == 1) & (df['RSI'] < 30), 1, 0)

    # 出場策略
    df['Stop_Loss'] = df['Close'] * 0.95
    df['Take_Profit'] = df['Close'] * 1.05
    df['Trend_Reversal'] = np.where((df['MA50'] < df['MA200']) & (df['MA50'].shift(1) > df['MA200'].shift(1)), -1, 0)
    df['Sell_Signal'] = 0
    df.loc[(df['Close'] < df['Stop_Loss']) | (df['Close'] > df['Take_Profit']) | (df['Trend_Reversal'] == -1), 'Sell_Signal'] = -1

    # 回測
    df['Position'] = 0
    df.loc[df['Buy_Signal'] == 1, 'Position'] = 1
    df.loc[df['Sell_Signal'] == -1, 'Position'] = 0
    df['Return'] = np.log(df['Close'] / df['Close'].shift(1))
    df['Strategy_Return'] = df['Return'] * df['Position'].shift(1)

    # 評估策略表現
    strategy_return = df['Strategy_Return'].cumsum()
    buy_hold_return = df['Return'].cumsum()
    plt.plot(strategy_return, label='Strategy Return')
    plt.plot(buy_hold_return, label='Buy and Hold Return')
    plt.legend(loc='upper left')
    plt.title('Strategy Performance')
    plt.show()
```

```python
def rebalance(context, data):
    today = data.current_session.date()
    todays_universe = context.tickers
    
    hist_close = data.history(todays_universe, 'close', bar_count=50, frequency='1d')
    hist_volume = data.history(todays_universe, 'volume', bar_count=50, frequency='1d')
    
    for asset in todays_universe:
        h_close = hist_close.unstack()[asset]
        h_volume = hist_volume.unstack()[asset]
        h_std = hist_close.unstack().pct_change().iloc[-20:].std()[asset]*100
        
        if asset in context.portfolio.positions:
            p = context.portfolio.positions[asset]
            
            if p.amount > 0:  # 長倉
                if h_close.iloc[-1] < h_close.iloc[-20]:  # 止損
                    order_target(asset, 0)
                elif h_close.iloc[-1] > h_close.iloc[-10] * (1 + context.take_profit):  # 獲利止損
                    order_target(asset, 0)
                elif h_close.iloc[-1] < h_close.iloc[-50]:  # 趨勢反轉
                    order_target(asset, 0)
                    
        else:  # 無倉位
            if asset == symbol('IR0001'): continue
            
            volume_to_trade = position_size(context.portfolio.portfolio_value, h_std, 1)
            
            # 等待回調買入
            if (h_close.iloc[-1] > h_close.iloc[-50]) and (h_close.iloc[-1] < h_close.iloc[-20]):
                order_value(asset, volume_to_trade)
                
            # 分批買入
            elif (h_close.iloc[-1] > h_close.iloc[-50]) and (h_close.iloc[-1] > h_close.iloc[-20]):
                order_value(asset, volume_to_trade / 2)
```
